```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from scipy import sparse
```

```python
ds = pd.read_csv("../data/tourist_attractions.csv")
ds.info()

tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0, stop_words='english')
tfidf_matrix_name = tf.fit_transform(ds['name'])
tfidf_matrix_region = tf.fit_transform(ds['region'])
tfidf_matrix_locality = tf.fit_transform(ds['locality'].fillna('None'))
combined_matrix = sparse.hstack((tfidf_matrix_name, tfidf_matrix_region, tfidf_matrix_locality))

cosine_similarities = linear_kernel(combined_matrix, combined_matrix)
print(cosine_similarities)
```

```python
import numpy as np
import matplotlib.pyplot as plt

results = {}

for idx, row in ds.iterrows():
    similar_indices = cosine_similarities[idx].argsort()[:-100:-1]
    similar_items = [(cosine_similarities[idx][i], ds['id'][i]) for i in similar_indices]

    results[row['id']] = similar_items[1:]

print(results[1])
print('done!')
```

```python
def item(id):
    return ds.loc[ds['id'] == id]['name'].tolist()[0].split(' - ')[0]

# Just reads the results out of the dictionary.
def recommend(item_id, num):
    print("Recommending " + str(num) + " products similar to " + item(item_id) + "...")
    print("-------")
    recs = results[item_id][:num]
    for rec in recs:
        print("Recommended: ID=" + str(rec[1]) + "; " + item(rec[1]) + " (score:" + str(rec[0]) + ")")
```

```python
recommend(item_id=2, num=5)
```
